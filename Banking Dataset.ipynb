##### Splitting the X and y

```python
data_X = data.drop(['Networth_Next_Year_1'], axis = 1)
data_y = data['Networth_Next_Year_1']
```

##### Dropping columns : Num(all values are unique), Networth_Next_Year(already converted to categorical column), Deposits_accepted_by_commercial_banks(100% null values)

```python
data_X.drop(['Num', 'Networth_Next_Year', 'Deposits_accepted_by_commercial_banks'], axis = 1, inplace = True)
```

##### In the dataset; columns(Other_income, Deferred_tax_liability, Contingent_liabilities, Investments, PE_on_BSE) consists null values >30%; hence dropping it as imputing such gaps may introduce biasness in the data.

```python
drop_cols = data_X.columns[(data_X.isnull().mean()*100) >= 30]
print(drop_cols.tolist())
```

```python
data_X.drop(drop_cols, axis = 1, inplace = True)
```

##### Outlier Checking

```python
data_1 = pd.concat([data_X, data_y], axis = 1)
data_1.head(1)
```

##### Feature Scaling and Imputing the null values

```python
predictors = data_1.drop('Networth_Next_Year_1', axis = 1)
response = data_1['Networth_Next_Year_1']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_predictors = pd.DataFrame(scaler.fit_transform(predictors), columns = predictors.columns)
scaled_predictors.head(1)
```

```python
data_2 = pd.concat([scaled_predictors, response], axis = 1)
data_2.head(1)
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors = 5)
imputed_array = imputer.fit_transform(data_2)
```

```python
data_imputed = pd.DataFrame(imputed_array, columns = data_2.columns)
data_imputed.head(1)
```

```python
data_imputed.isnull().sum()
```

##### Importing all the necessary models/algorithms, evaluation metrics, hyperparameter tuning

```python
from sklearn.model_selection import cross_val_score, GridSearchCV, KFold, RandomizedSearchCV, train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, r2_score, recall_score, f1_score, precision_score, roc_curve, auc
import statsmodels.formula.api as snf
from sklearn import tree
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_imputed.drop('Networth_Next_Year_1', axis = 1), data_imputed['Networth_Next_Year_1'], 
                                                    test_size = 0.25, random_state = 42)
```

##### This is imbalanced dataset since there is huge difference between occurence of both classes. Hence, using SMOTE technique; we will balance the imbalanced dataset

```python
import imblearn
from imblearn.over_sampling import SMOTE
```

```python
data_imputed['Networth_Next_Year_1'] = np.where(data_imputed['Networth_Next_Year_1'] == 0.0, 0, 1)
```

```python
sm = SMOTE(random_state = 42)
X_bal, y_bal = sm.fit_resample(X_train, y_train)
```

```python
pred = data_imputed.drop('Networth_Next_Year_1', axis = 1)
res = data_imputed['Networth_Next_Year_1']
```

```python
X_smote, y_smote = sm.fit_resample(pred, res)
```

```python
y_bal.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_smote, y_smote, test_size = 0.25, random_state = 42)
```

##### Model Building

```python
# Logistic Regression
model_lor = LogisticRegression()
model_lor = model_lor.fit(X_train, y_train)
y_pred_lor = model_lor.predict(X_test)
accuracy_lor = accuracy_score(y_test, y_pred_lor)

# Random Forest
model_rf = RandomForestClassifier()
model_rf = model_rf.fit(X_train, y_train)
y_pred_rf = model_rf.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)

# KNN
model_knn = KNeighborsClassifier()
model_knn = model_knn.fit(X_train, y_train)
y_pred_knn = model_knn.predict(X_test)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
```

```python
accuracy_percent = pd.DataFrame({"Model" : ["Logistic Regression", "Random Forest", "KNN"], "Accuracy" : [accuracy_lor, accuracy_rf, accuracy_knn]})
accuracy_percent
```

##### As per the table : Model vs Accuracy, we can see that the best accuracy we are receiving is from Random Forest model.
